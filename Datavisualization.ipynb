```
#first to import matplotlib
import matplotlib.pyplot as plt
```

```
distance=[10,23,50,45]  #distance in KM
time=[7,12,15,23]       #time in minutes
```

```
plt.xlabel("distance")     #for X axis
plt.ylabel("time")         #for Y axis
plt.plot(distance,time,label="Line")    #for line graph
plt.scatter(distance,time,label='Dot') #for scatter graph
plt.grid(c='g')            #for Grids
plt.legend()               #To Show labels
plt.title("Line Graph")    #Title of Graph
plt.show()                 #To show Graph
```

```
#This is cricket data
players=["Virat","Dhoni","Rohit"]
runs=[120,90,150]
```

```
plt.xlabel("players")
plt.ylabel("runs")
#plt.bar(players,runs)
plt.pie(runs,labels=players)
plt.grid(c='g')
plt.legend()
plt.show()
```

```
plt.xlabel("players")
plt.ylabel("runs")
#plt.bar(players,runs)
plt.pie(runs,labels=players,explode=(0.1,0.1,0.2))
plt.grid(c='g')
plt.legend()
plt.show()
```

```
plt.xlabel("players")
plt.ylabel("runs")
#plt.bar(players,runs)
plt.pie(runs,labels=players,explode=(0.1,0.1,0.2),shadow=True,autopct='%1.1f%%')
plt.grid(c='g')
plt.legend()
plt.show()
```
